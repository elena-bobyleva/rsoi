```python
from xmlrpc.server import SimpleXMLRPCServer
from xmlrpc.server import SimpleXMLRPCRequestHandler

import datetime
import os
import sqlite3
import pandas as pd

class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_paths = ('/RPC2',)

server = SimpleXMLRPCServer(("localhost", 8018), 
                            requestHandler=RequestHandler)
# def create_log_table():
#     # Создаем подключение к базе данных (файл my_database.db будет создан)
#     connection = sqlite3.connect('log.db')
#     cursor = connection.cursor()
    

#     # Сохраняем изменения и закрываем соединение
#     connection.commit()
#     connection.close()
#     return True

# server.register_function(create_log_table, 'create_log_table')

# Добавление строки в лог
def add_log(sname, date_time, proc_time):
    connection = sqlite3.connect('log.db')
    cursor = connection.cursor()
    
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Log (
    id INTEGER PRIMARY KEY,
    sname TEXT NOT NULL,
    date_log TEXT NOT NULL,
    proc_time INTEGER
    )
    ''')
    connection.commit()
    
    cursor.execute('INSERT INTO Log (sname, date_log, proc_time) VALUES (?, ?, ?)', (sname, date_time, proc_time))
    
    connection.commit()
    connection.close()
    
    return True

server.register_function(add_log, 'add_log')

def get_data_log():
    connection = sqlite3.connect('log.db')
    cursor = connection.cursor()
    cursor.execute('SELECT sname, date_log, proc_time FROM Log')
    results = cursor.fetchall()
    connection.commit()
    connection.close()
    return results
server.register_function(get_data_log, 'get_data_log')

def get_data_log_by_type(type_name):
    connection = sqlite3.connect('log.db')
    cursor = connection.cursor()
    cursor.execute('SELECT * FROM Log WHERE sname = ?', (type_name,))
    results = cursor.fetchall()
    connection.commit()
    connection.close()
    return results
server.register_function(get_data_log_by_type, 'get_data_log_by_type')

def get_data_log_by_time(start_time, end_time):
    connection = sqlite3.connect('log.db')
    cursor = connection.cursor()
    cursor.execute('SELECT sname, date_log, proc_time FROM Log WHERE date_log >= ? AND date_log <= ?', (start_time, end_time))
    results = cursor.fetchall()
    connection.commit()
    connection.close()
    return results
server.register_function(get_data_log_by_time, 'get_data_log_by_time')

def get_data_log_by_proc_time(start_time, end_time):
    connection = sqlite3.connect('log.db')
    cursor = connection.cursor()
    cursor.execute('SELECT sname, date_log, proc_time FROM Log WHERE proc_time >= ? AND proc_time <= ?', (start_time, end_time))
    results = cursor.fetchall()
    connection.commit()
    connection.close()
    return results
server.register_function(get_data_log_by_proc_time, 'get_data_log_by_proc_time')

print ("Listening on port 8018...")
server.serve_forever()
```
