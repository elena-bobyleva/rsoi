```python
from xmlrpc.server import SimpleXMLRPCServer
from xmlrpc.server import SimpleXMLRPCRequestHandler
import time
import xmlrpc.server
import datetime

class RequestHandler(SimpleXMLRPCRequestHandler):
    rpc_paths = ('/RPC2',)

server = SimpleXMLRPCServer(("localhost", 8028), requestHandler=RequestHandler)
#server.register_introspection_functions()

server_server = xmlrpc.client.ServerProxy("http://localhost:8008")

stats_server = xmlrpc.client.ServerProxy("http://localhost:8018")

# Добавление в лог через сервер
def add_log(sname, proc_time):
    try:
        stats_server.add_log(sname, datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S"), proc_time)
    except Exception:
        print('Failed to add data to log. The statistics service server is not available.')
    return True

def ping():
    start = time.time()
    res = server_server.ping()
    end = time.time()
    add_log('ping', end-start)
    return res
server.register_function(ping, 'ping')

# Время сервера
def now():
    start = time.time()
    res = server_server.now()
    end = time.time()
    add_log('now', end-start)
    return res
server.register_function(now, 'now')

# Отображение строкового вида, типа и значений
def show_type(arg):
    start = time.time()
    res = server_server.type(arg)
    end = time.time()
    add_log('type', end-start)
    return res
server.register_function(show_type, 'type')

# Сумма
def test_sum(a, b):
    start = time.time()
    res = server_server.sum(a, b)
    end = time.time()
    add_log('sum', end-start)
    return res
server.register_function(test_sum, 'sum')

# Степень
def test_pow(a, b):
    start = time.time()
    res = server_server.pow(a, b)
    end = time.time()
    add_log('pow', end-start)
    return res
server.register_function(test_pow, 'pow')

# Проверка нахождения клиента в черном списке c использованием Pandas Data Frame
def black_list_check(sname):
    start = time.time()
    res = server_server.black_list_check(sname)
    end = time.time()
    add_log('black_list_check', end-start)
    return res
server.register_function(black_list_check, 'black_list_check')

# Проверка на наличие человека в черном списке с проверкой по ФИО и дате рождения
def black_list_check_fio(sname, name, patronym, birth):
    start = time.time()
    res = server_server.black_list_check_fio(sname, name, patronym, birth)
    end = time.time()
    add_log('black_list_check_fio', end-start)
    return res
server.register_function(black_list_check_fio, 'black_list_check_fio')

# Бинарная передача данных
def send_back_binary(bin_data):
    start = time.time()
    res = server_server.back_binary(bin_data)
    end = time.time()
    add_log('back_binary', end-start)
    return res
server.register_function(send_back_binary, 'back_binary')
    
# Инверсия цвета
# На вход изображение RGB размерности (M, N, 3) со значениями 0-255
# или монохромные изображения размерностью (M, N, 1)
def send_back_inversion(bin_data):
    start = time.time()
    res = server_server.color_inversion(bin_data)
    end = time.time()
    add_log('color_inversion', end-start)
    return res  
server.register_function(send_back_inversion, 'color_inversion')

# Разворот изображения относительно вертикали для цветных и монохромных изображений
def display_image_vertically(bin_data):
    start = time.time()
    res = server_server.display_image_vertically(bin_data)
    end = time.time()
    add_log('display_image_vertically', end-start)
    return res 
server.register_function(display_image_vertically, 'display_image_vertically')

# Бинаризация изображения по задаваемому порогу 1-255 для цветных и монохромных изображений
def binarize_image_by_threshold(bin_data, threshold):  
    start = time.time()
    res = server_server.binarize_image_by_threshold(bin_data, threshold)
    end = time.time()
    add_log('binarize_image_by_threshold', end-start)
    return res 
server.register_function(binarize_image_by_threshold, 'binarize_image_by_threshold')

def get_data_log():
    try:
        res = stats_server.get_data_log()
    except Exception:
        print('The statistics service server is not available.')
    return res
server.register_function(get_data_log, 'get_data_log')

def get_data_log_by_type(type_name):
    try:
        res = stats_server.get_data_log_by_type(type_name)
    except Exception:
        print('The statistics service server is not available.')
    return res
server.register_function(get_data_log_by_type, 'get_data_log_by_type')

def get_data_log_by_time(start_time, end_time):
    try:
        res = stats_server.get_data_log_by_time(start_time, end_time)
    except Exception:
        print('The statistics service server is not available.')
    return res
server.register_function(get_data_log_by_time, 'get_data_log_by_time')

def get_data_log_by_proc_time(start_time, end_time):
    try:
        res = stats_server.get_data_log_by_proc_time(start_time, end_time)
    except Exception:
        print('The statistics service server is not available.')
    return res
server.register_function(get_data_log_by_proc_time, 'get_data_log_by_proc_time')

print("Listening on port 8028...")
server.serve_forever()
    
```
